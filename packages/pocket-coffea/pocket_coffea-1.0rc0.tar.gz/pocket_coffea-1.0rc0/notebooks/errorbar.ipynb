```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,11)
y = 2*x
xerr = 0.5
yerr = 0.1*y
```

```python
fig, ax = plt.subplots(1,1, figsize=[16,10])
lines = plt.errorbar(x, y, yerr, xerr, marker='.', linestyle='none');
line = lines[-1][0]
line2 = lines[-1][1]
line.set_linestyle('--')
line2.set_linestyle(':')
line2.set_linewidth(0)
```
