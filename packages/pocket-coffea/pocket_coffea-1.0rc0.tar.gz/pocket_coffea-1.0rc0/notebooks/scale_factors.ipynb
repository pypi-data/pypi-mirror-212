```python
import numpy as np
import correctionlib
import matplotlib.pyplot as plt
```

```python
jsonFile = "/cvmfs/cms.cern.ch/rsync/cms-nanoAOD/jsonpog-integration/POG/EGM/2018_UL/electron.json.gz"
evaluator      = correctionlib.CorrectionSet.from_file(jsonFile)
evaluator_core = correctionlib._core.CorrectionSet.from_file(jsonFile)
```

```python
for k in evaluator.keys():
    print(k)
evaluator
```

```python
pt = np.linspace(20, 300, 281)
eta = np.ones_like(pt)
sf      = evaluator['UL-Electron-ID-SF'].evaluate("2018", "sf", "RecoAbove20", eta, pt)
sf_core = evaluator['UL-Electron-ID-SF'].evaluate("2018", "sf", "RecoAbove20", eta, pt)
fig, (ax1, ax2) = plt.subplots(1,2, figsize =(16, 6))
ax1.plot(pt, sf, label='highlevel')
ax1.plot(pt, sf_core, label='core')
ax1.legend()
ax2.plot(pt, sf_core - sf)
ax2.set_title("core - highlevel");
```
